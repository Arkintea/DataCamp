```python
features = df[['FirstServeReturnPointsWon']]
winnings = df['Winnings']
X_train, X_test, y_train, y_test = train_test_split(features, winnings, test_size = 0.2, random_state = 1)
model = LinearRegression()
model.fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
y_predict = model.predict(X_test)
plt.scatter(y_test, y_predict, alpha=0.4)
plt.title('FirstServeReturnPointsWon vs Winnings')
plt.xlabel('Winnings')
plt.ylabel('FirstServeReturnPointsWon')
plt.show()
```
